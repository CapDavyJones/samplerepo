```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Prince\Desktop\DataRelated\healthcare_messy_data.csv")
```

```python
df
```

```python
# Strip leading and trailing spaces from string columns 
for column in df.select_dtypes(include=['object']).columns: 
    df[column] = df[column].str.strip()
```

```python
# Correct the Age column 
df['Age'] = df['Age'].replace('forty', 40)  # Replace 'forty' with 40
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert to numeric, coercing errors to NaN

# Fill NaN values in the Age column before converting to int
df['Age'] = df['Age'].fillna(0).round(0).astype(int)  # Remove decimals and convert to int
```

```python
# Correct the Blood Pressure column 
df['Blood Pressure'] = pd.to_numeric(df['Blood Pressure'], errors='coerce')
 
 # Correct the Cholesterol column 
df['Cholesterol'] = pd.to_numeric(df['Cholesterol'], errors='coerce')  # Convert to numeric, coercing errors to NaN

# Fill NaN values in the Cholesterol column before converting to int
df['Cholesterol'] = df['Cholesterol'].fillna(0).round(0).astype(int)  # Remove decimals and convert to int
```

```python
# Standardise the Visit Date column 
df['Visit Date'] = pd.to_datetime(df['Visit Date'], errors='coerce')
```

```python
# Fill NaN values with a default value or strategy 
df['Age'].fillna(df['Age'].mean(), inplace=True) 
df['Cholesterol'].fillna(df['Cholesterol'].mean(), inplace=True) 
 
 # Fill NaN values for categorical columns with a placeholder or most frequent value 
df['Gender'].fillna('Unknown', inplace=True) 
df['Condition'].fillna(df['Condition'].mode()[0], inplace=True) 
df['Medication'].fillna(df['Medication'].mode()[0], inplace=True) 
df['Blood Pressure'].fillna('120/80', inplace=True) 
df['Email'].fillna('no_email@example.com', inplace=True) 
df['Phone Number'].fillna('000-000-0000', inplace=True)
```

```python
# Save the cleaned data to a new CSV file 
df.to_csv('cleaned_healthcare_data.csv', index=False)
```
