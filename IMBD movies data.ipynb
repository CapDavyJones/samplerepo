```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Prince\Desktop\DataRelated\messy_IMDB_dataset.csv", encoding='ISO-8859-1', sep=None, engine='python')
```

```python
# Inspect the dataset 

df.head()

df.info()

df.describe(include='all')
```

```python
df.rename(columns={
    'IMBD title ID': 'IMDB_title_ID',
    'Original titlÊ': 'Original_title',
    'Release year': 'Release_year',
    'Genrë¨': 'Genre',
    'Duration': 'Duration',
    'Country': 'Country',
    'Content Rating': 'Content_Rating',
    'Director': 'Director',
    'Income': 'Income',
    'Votes': 'Votes',
    'Score': 'Score'
}, inplace=True)
```

```python
# Handle missing values 

df['Duration'].fillna('Unknown', inplace=True)

df['Content_Rating'].fillna('Not Rated', inplace=True)

# Drop rows with critical missing data

df.dropna(subset=['IMDB_title_ID', 'Original_title', 'Release_year'], inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()  # Remove any leading or trailing whitespace
```

```python
# Convert 'Release_year' to datetime
df['Release_year'] = pd.to_datetime(df['Release_year'], errors='coerce')

# Clean and convert 'Income' to numeric
df['Income'] = pd.to_numeric(df['Income'].replace('[\$,]', '', regex=True), errors='coerce')

# Clean and convert 'Votes' to numeric
# Ensure 'Votes' exists in the DataFrame
if 'Votes' in df.columns:
    df['Votes'] = pd.to_numeric(df['Votes'].replace('[\.,]', '', regex=True), errors='coerce')
else:
    print("Column 'Votes' does not exist in the DataFrame.")
```

```python
# Clean and standardize the 'Score' column 

df['Score'] = pd.to_numeric(df['Score'].replace('[^0-9.]', '', regex=True), errors='coerce')

# Normalise any scores greater than 10

df['Score'] = df['Score'].apply(lambda x: x/10 if x > 10 else x)
```
